```python
with open("access.csv", "r") as fichier:
    data = fichier.readlines()

data = list(map(lambda x: x.split(";"), data))[1:]

stat = {}
for d in data:
    if d[4] in stat:
        statCodeStatus = stat[d[4]]

        statCodeStatus["TOTAL"] += 1

        if d[2] in statCodeStatus:
            statCodeStatus[d[2]] += 1
        else:
            statCodeStatus[d[2]] = 1
    else:
        stat[d[4]] = {"TOTAL": 1, d[2]: 1}

# triage des codeStatus
stat_sort = {
    k: v
    for k, v in sorted(stat.items(), key=lambda item: item[1]["TOTAL"], reverse=True)
}

# triage des Methode post
for key, value in stat_sort.items():
    stat_sort[key] = {
        k: v for k, v in sorted(value.items(), key=lambda item: item[1], reverse=True)
    }

print(stat_sort)
```

```python
total_general = sum(value['TOTAL'] for value in stat_sort.values())

totaux_par_code = {}

for code, values in stat_sort.items():
    total_code = values['TOTAL']

    total_get = values.get('GET', 0)
    total_post = values.get('POST', 0)
    total_head = values.get('HEAD', 0)

    pourcentage_get = (total_get / total_code) * 100
    pourcentage_post = (total_post / total_code) * 100
    pourcentage_head = (total_head / total_code) * 100

    totaux_par_code[code] = {
        'TOTAL': total_code,
        'GET': total_get,
        'POST': total_post,
        'HEAD': total_head,
        'Pourcentage GET': pourcentage_get,
        'Pourcentage POST': pourcentage_post,
        'Pourcentage HEAD': pourcentage_head
    }

for code, values in totaux_par_code.items():
    print(f"Code de statut {code}:")
    print(f"GET: {values['GET']} ({values['Pourcentage GET']:.2f}%)")
    print(f"POST: {values['POST']} ({values['Pourcentage POST']:.2f}%)")
    print(f"HEAD: {values['HEAD']} ({values['Pourcentage HEAD']:.2f}%)")
    print(f"Total GET, POST et HEAD: {values['TOTAL']}")
    print()
```

```python
heure_counts = {}
for d in data:
    if len(d) > 1 and d[1]:
        heure = d[1].split()[1].split(":")[0]
        if heure in heure_counts:
            heure_counts[heure] += 1
        else:
            heure_counts[heure] = 1

heure_pointe = max(heure_counts, key=heure_counts.get)
nombre_requetes_pointe = heure_counts[heure_pointe]

print(f"L'heure de pointe de la journée, où il y a le plus grand nombre de requêtes, est: {heure_pointe}h avec {nombre_requetes_pointe} requêtes.")
```

```python
statistiques_par_jour = {}

for d in data:

    if d and d[0]:

        date_heure = d[1]
        date = date_heure.split()[0]
        code_statut = d[4]

        if date in statistiques_par_jour:
            if code_statut in statistiques_par_jour[date]:
                statistiques_par_jour[date][code_statut] += 1
            else:
                statistiques_par_jour[date][code_statut] = 1
        else:
            statistiques_par_jour[date] = {code_statut: 1}
            
for date, statistiques in statistiques_par_jour.items():
    print(f"Date : {date}")
    total_requetes = sum(statistiques.values())
    print(f"Total de requêtes : {total_requetes}")
    for code_statut, nombre_requetes in statistiques.items():
        print(f"Code de statut {code_statut} : {nombre_requetes}")
    print()
```
